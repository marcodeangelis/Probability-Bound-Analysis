# Tutorial Confidence Boxes

Confidence boxes (cboxes) are imprecise generalizations of confidence distributions. They encode frequentist confidence intervals, at every confidence level, for parameters of interest and thereby characterize the inferential uncertainty about distribution parameters estimated from sparse or imprecise sample data. They have a purely frequentist interpretation that makes them useful in engineering because they offer a guarantee of statistical performance through repeated use. Unlike traditional confidence intervals which cannot usually be propagated through mathematical calculations, c-boxes can be used in calculations using the standard methods of probability bounds analysis and yield results that also admit the same confidence interpretation. This means that analysts using them can now literally compute with confidence.

## Cbox for the Binomial Probability

A Bernoulli random variable has only two possible values, perhaps designated {failure, success}, or more conveniently $[0, 1]$. A binomial random variable is a random variable whose value is a count of Bernoulli successes observed over $n > 0$ independent identical trials, each of which has the same probability $p$ of success, which produces $k$ successes from those $n$ trials (where $0 ≤ k ≤ n$). A fundamental problem in risk analysis and statistics generally is to characterize what can be inferred about $p$ from observing $k$ successes out of $n$ trials, under the assumption that the trials are independent and the binomial probability $p$ is fixed across the trials.

Balch (2012) offers a c-box solution to this problem with the following approximation:

$$ p \approx beta(k, n-k+1), beta(k+1, n-k)$$


In this tutorial, I will demostrate how to compute cboxes, compute focal elements, compute confidence intervals and perfrom arithmetic opertation on cboxes.

```python
from pba import *
```

Now let's construct a cbox from $k$ successes out of $n$ trials. In this example, $k$ = 2 and $n$ = 10.

```python
cb = cBox(2, 10)
```

Now let's visualise the cbox using a plot function.

```python
plotcBox(cb)
```

Now, let's compute 10 focal elements ($\alpha$-slices) of the pbox above and visualise them:

```python
computeFocalElements(cb, npoints=10, show_plot = True)
```

We can also change the number of focal elements in the plot. For example, npoints = 5.

```python
computeFocalElements(cb, npoints = 5, show_plot = True)
```

We can also compute the 100($\beta$-$\alpha$)%  confidence level for a cbox.

```python
computeConfidenceInterval(cBox(1,2), show_plot = True)
```

The defaut confidence level is $\alpha$ = 0.05 and $\beta$ = 0.95. We can compute for diferent levels by changing the values of $\alpha$ and $\beta$.

```python
computeConfidenceInterval(cBox(2,4), show_plot=True, alpha_level=0.1, beta_level=0.8)
```

Now, let's add, subtract, multiply and divide cboxes. Remember we constructed a cbox (cb) eailer, now let's construct another cbox:

```python
cb2 = cBox(1, 3)
```

```python
plotcBox(cb2)
```

Now let's do cb3= cb + cb2

```python
cb3 = addCbox(cb, cb2, show_plot = True)
```

Note that adding 2 beta distributions dosen't give you another beta distribution

```python
cb4 = subtractCbox(cb3, cb, show_plot = True)
```

```python
cb5 = multiplyCbox(cb, cb2, show_plot = True)
```

```python
cb6 = divideCbox(cBox(1,2),cBox(2,3), show_plot = True)
```
